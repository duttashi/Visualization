This script is adapted from a stackoverflow question(https://stackoverflow.com/questions/2895319/how-to-add-texture-to-fill-colors-in-ggplot2). The question was, "How to add texture to fill colors in ggplot2?"

```r
library(ggplot2)
ggplot(diamonds, aes(x=cut, y=price, group=cut))+
geom_boxplot(aes(fill=cut))+scale_fill_brewer(palette="OrRd")
```
